## Calculate features from Image data

These include:
* wwr_extracted: the WWR that was extracted using the outside area and the detected facade elements
* Solid Wall Area: with the WWR, the solid wall area for each wall is computed

```python
# paths
data_dir = "./../../data"
dataframes_dir = data_dir + "/dataframes"

# read data in
df2_walls = pd.read_csv(dataframes_dir + "/df2_walls_04_with_wwr.csv")
df2_walls
```


1. **WWR** is renamed to **wwr_extracted**


```python
# 1. WWR is renamed to **wwr_extracted**
df2_walls = df2_walls.rename(columns={'wwr': 'wwr_extracted'})

# 2. Remove rows with no WWR
df2_walls = df2_walls.loc[~df2_walls["wwr_extracted"].isna()]
df2_walls
```

### Solid Wall Area
The solid wall area is the total outside area multiplied by the complement of the wwr_corrected value.

```python
# solid wall area: outside area * (1-wwr)
df2_walls["solid_wall_area"] = (1 - df2_walls["wwr_extracted"]) * df2_walls["outside_area"]
```

```python
# save final dataframe
df2_walls.to_csv(dataframes_dir + "/df2_walls_05_with_image_factors.csv", index = False)
```
